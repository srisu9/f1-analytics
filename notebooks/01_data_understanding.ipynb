```python
drivers = pd.read_csv("../data/raw/drivers.csv")
constructors = pd.read_csv("../data/raw/constructors.csv")
races = pd.read_csv("../data/raw/races.csv")
results = pd.read_csv("../data/raw/results.csv")
qualifying = pd.read_csv("../data/raw/qualifying.csv")
circuits = pd.read_csv("../data/raw/circuits.csv")
```

```python
def explore_data(df, name):
    print("=" * 80)
    print(f"Dataset: {name}")
    print("=" * 80)

    print("\nShape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns)

    print("\nInfo:")
    df.info()

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nFirst 5 Rows:")
    display(df.head())

    print("\nSummary Statistics:")
    print("Numerical Columns")
    display(df.describe())

    print("\nCategorical Columns")
    display(df.describe(include='object'))

    print("\n\n")
```

```python
explore_data(drivers, "Drivers")

explore_data(constructors, "Constructors")

explore_data(races, "Races")

explore_data(results, "Results")

explore_data(qualifying, "Qualifying")

explore_data(circuits, "Circuits")
```
